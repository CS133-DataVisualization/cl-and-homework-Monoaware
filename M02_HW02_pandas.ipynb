#### Question 5: ####
What is the title of the wine that has the highest points-to-price ratio in the dataset?

The points-to-price ratio can be calculuated by dividing the points value in each row by the price value in each row. We can find the highest by applying pandas.DataFrame.max to the points-to-price ratio column.

```python
# Calculate the points-to-price ratio:
points_to_price_ratio = wine['points'] / wine['price']

# Locate the position of the maximum:
max_index = points_to_price_ratio.idxmax() 

# Query the name of the wine in the index of the maximum ratio:
highest_wine = wine['title'].iloc[max_index] # This only works when we haven't modified the index in any way (which we haven't... yet).
# More robust query:
highest_wine_robust = wine.loc[max_index, 'title']

print("Wine with the highest points-to-price ratio in dataset:", highest_wine)
print("Testing robustness:", highest_wine_robust)
```

**Answer:** The title of the wine with the highest points-to-price ratio in the dataset is Bandit NV Merlot (California).

#### Question 6: ####
Create a series `flavor_counts` that contains two values: the number of wines that has the word "tart" in the `description` column and the number of wines that has the word "berries" in the `description` column. The index of the Series should be "Tart" and "Berries" for the corresponding values.

We can use pandas.Series.str.contains to help filter for specific description keywords. More specifically, we need to first select the description column, then pass in the regex to match with case sensitivity set to false. We cannot use literal string matching because that will match substrings too. Just in case there is no description (NaN), we instruct the function to ignore it. Afterwards, we need to apply the pandas.Series.sum function to count the number of descriptions containing matching keywords.

```python
# Look for the word "tart", not case sensitive, not a substring of another word.
# Same applies to the word "berries"
flavor_counts = pd.Series({
    "Tart": wine['description'].str.contains(r"\btart\b", case=False, na=False, regex=True).sum(), # regex is True by default.
    "Berries": wine['description'].str.contains(r"\bberries\b", case=False, na=False, regex=True).sum()
})

flavor_counts
```

#### Question 7: ####
Let's convert the points into simple star ratings. A score of 90 or higher counts as 3 stars, a score of at least 80 but less than 90 is 2 stars. Any other score is 1 star.

Also, any wines from France should automatically get 3 stars, regardless of points.

Add this new column `star_ratings` to the dataframe with the number of stars for each wine in the dataset.

We'll need to build the star rating columns step-by-step.

```python
# Let's start with the default, 1 star:
wine['star_ratings'] = 1

# Now we can handle 2 star ratings using pandas.DataFrame.loc to select rows where our conditions are True, then select the given column.
wine.loc[wine['points'] >= 80, 'star_ratings'] = 2

# Repeat for other 3 stars.
wine.loc[wine['points'] >= 90, 'star_ratings'] = 3
wine.loc[wine['country'] == 'France', 'star_ratings'] = 3

wine.head()
```

#### Question 9: ####
What combination of countries and varieties are most common? Create a Series whose index is a MultiIndexof {country, variety} pairs. For example, a pinot noir produced in the US should map to {"US", "Pinot Noir"}. Sort the values in the Series in descending order based on wine count.

This is another situation where pandas.DataFrame.value_counts can be used. Alternatively, we can create the multi-index using pandas.DataFrame.groupby, then pandas.Series.sort_values.

```python
# Option 1: pandas.DataFrame.value_counts
country_variety_wine_counts = (wine[['country', 'variety']].value_counts())
country_variety_wine_counts.head(10)
```

```python
# Option 2: pandas.DataFrame.groupby into pandas.DataFrameGroupBy.size into pandas.Series.sort_values
country_variety_wine_counts_alternative = (wine.groupby(['country', 'variety']).size().sort_values(ascending=False))
country_variety_wine_counts_alternative.head(10)
```

#### Question 10 #####
Create a Series whose index is reviewers and whose values is the average score given out by that reviewer.  
*Hint:* You will need the `taster_name` and `points` columns.
